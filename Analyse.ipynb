```python
import pandas as pd
```

```python
# lecture du csv
df = pd.read_csv('gameandgrade new.csv')
```

```python
# Vérifie s'il y a des valeurs manquantes
print(df.isnull().any().any())  # True si au moins une valeur est manquante
```

```python
print(df.duplicated().any())  # True si au moins une ligne est dupliquée
```

```python
# Afficher les doublons
print(df[df.duplicated()])
```

```python
df.info()
```

```python
df['Grade'] = df['Grade'].astype(str).str.strip()           # Supprime les espaces
df['Grade'] = df['Grade'].str.replace(',', '.', regex=False)  # Remplace les virgules par des points
```

```python
df['Grade'] = pd.to_numeric(df['Grade'], errors='coerce') # errors='coerce' transforme les valeurs non convertibles en NaN (valeurs manquantes) | ensuite on peut remplacer Nan par une valeur par défaut avec fillna(0)
```

```python
# Vérifie s'il y a des valeurs manquantes
print(df.isnull().any().any())  # True si au moins une valeur est manquante
```

```python
lignes_incompletes = df[df.isnull().any(axis=1)]
print(lignes_incompletes)
```

```python
df["Grade"].dtype
df.head
```

```python
# replacement de la valeur NaN par sa bonne valeur (92)
df['Grade'] = df['Grade'].fillna(92)
df.loc[363]
```

```python
# Replacer la colonne percentage de la ligne 363 par sa bonne valeur
df.at[363, 'percentage'] = "9200,00%"
df.loc[363]
```

```python
#transformer toute la colonne en type pourcentage
df['percentage'] = (
    df['percentage']
    .astype(str)                          # Convertit en chaîne
    .str.strip()                          # Supprime les espaces
    .str.replace('%', '', regex=False)   # Supprime le symbole %
    .str.replace(',', '.', regex=False)  # Remplace les virgules par des points
)
df['percentage'] = pd.to_numeric(df['percentage'], errors='coerce')
```

```python
df['percentage'] = df['percentage'] / 100
print(df['percentage'])
```
